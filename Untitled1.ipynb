```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo Excel
file_path = "Servicios Prestados.xlsx"  # Reemplaza con la ruta correcta
df = pd.read_excel(file_path)

# Convertir la columna "Fecha" a formato datetime
df["Fecha"] = pd.to_datetime(df["Fecha"])

# Extraer el mes y año
df["Mes"] = df["Fecha"].dt.to_period("M")

# Calcular ingresos totales por mes
ingresos_mensuales = df.groupby("Mes")["Valor_Pagado"].sum().reset_index()

# Convertir la columna "Mes" a string para la visualización
ingresos_mensuales["Mes"] = ingresos_mensuales["Mes"].astype(str)

# Mostrar los ingresos mensuales
print(ingresos_mensuales)

# Graficar los ingresos mensuales
plt.figure(figsize=(10, 5))
plt.bar(ingresos_mensuales["Mes"], ingresos_mensuales["Valor_Pagado"], color="skyblue")
plt.xlabel("Mes")
plt.ylabel("Ingresos Totales ($)")
plt.title("Ingresos Mensuales de la Veterinaria")
plt.xticks(rotation=45)
plt.show()

```

```python
# Encontrar el mes con mayores ingresos
mes_mayor_ingreso = ingresos_mensuales.loc[ingresos_mensuales["Valor_Pagado"].idxmax()]
print(f"📈 Mes con mayor ingreso: {mes_mayor_ingreso['Mes']} (${mes_mayor_ingreso['Valor_Pagado']})")

# Encontrar el mes con menores ingresos
mes_menor_ingreso = ingresos_mensuales.loc[ingresos_mensuales["Valor_Pagado"].idxmin()]
print(f"📉 Mes con menor ingreso: {mes_menor_ingreso['Mes']} (${mes_menor_ingreso['Valor_Pagado']})")
```
